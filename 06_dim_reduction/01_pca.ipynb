```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# 붓꽃 데이터셋 불러오기 (4차원)
iris = load_iris()
X = iris.data
y = iris.target

# t-SNE 모델 생성 및 학습
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# 2차원으로 축소된 데이터 시각화
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.title('t-SNE 시각화 (Iris 데이터)')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['target'] = iris_data.target
iris_df.head()
```

```python
iris_data.target_names
```

```python
iris_df[iris_df['target'] == 0]
```

```python
iris_df[iris_df['target'] == 0]['sepal length (cm)']
```

```python
# 주요 차원 축소 기법
#  1. 특성 선택(Feature Selection)
#  2. 특성 추출(Feature Extraction)

# 특성 선택을 통한 시각화 (2개의 특성만 선택하여 시각화)
markers = ['^', 's', 'o']

for i, marker in enumerate(markers):
    X = iris_df[iris_df['target'] == i]['sepal length (cm)']
    y = iris_df[iris_df['target'] == i]['sepal width (cm)']
    plt.scatter(X, y, marker=marker, label=iris_data.target_names[i])

plt.legend()
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
# PCA 주성분 분석
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # 2차원으로 축소

X = iris_df.iloc[:, :-1] # 독립 변수(특성 데이터)만 모아서 X 데이터 생성

# 차원 축소 진행
pca.fit(X) # 주성분 분석 수행 → 축을 찾기 위한 데이터 분석 과정
iris_pca = pca.transform(X) # 주성분 분석 결과 데이터 변환
# X.shape ,iris_pca.shape # 2차원의 배열로 변함, iris_pca.shape: (150, 2)
# iris_pca #[[-2.68412563, 0.31939725],...] 1번째 주성분에 대한 변환 결과, 2번째 주성분에 대한 변환 결과

iris_pca_df = pd.DataFrame(data=iris_pca, columns=['pca_col1', 'pca_col2'])
iris_pca_df['target'] = iris_df.target

# 축소한 차원으로 PCA 통한 시각화
markers = ['^', 's', 'o']

for i, marker in enumerate(markers):
    X = iris_pca_df[iris_pca_df['target'] == i]['pca_col1']
    y = iris_pca_df[iris_pca_df['target'] == i]['pca_col2']
    plt.scatter(X, y, marker=marker, label=iris_data.target_names[i])

plt.legend()
plt.xlabel('pca_col1')
plt.ylabel('pca_col2')
plt.show()
```

```python
# PCA를 통해 생성된 주성분들이 실제 데이터의 총 분산에 얼마만큼의 비율을 나타내는지
pca.explained_variance_ratio_ # array([0.92461872, 0.05306648])
```

```python
# 원본 데이터 학습 및 평가
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

lr_clf = LogisticRegression(max_iter=1000)
scores = cross_val_score(lr_clf, iris_data.data, iris_data.target, cv=5)
# scores # array([0.96666667, 1.        , 0.93333333, 0.96666667, 1.        ]) → 5번의 교차 검증
print("원본 데이터 평가:", np.mean(scores))
```

```python
# PCA 데이터 학습 및 평가

lr_clf = LogisticRegression(max_iter=1000)
scores = cross_val_score(lr_clf, iris_pca_df[['pca_col1', 'pca_col2']], iris_pca_df['target'], cv=5)

print("PCA 데이터 평가:", np.mean(scores))
```

### PCA 과일 데이터

```python
fruits = np.load('./data/fruits_300.npy')
fruits.shape # (300, 100, 100) 300개의 이미지 * 100행 * 100열
# 흑백 이미지로 데이터의 픽셀 값을 0-255 사이의 값을 갖는다.
```

```python
# 이미지 시각화 함수
def draw_fruits(arr, ratio=1):
    N = len(arr) # 이미지 개수
    rows = int(np.ceil(N / 10)) # 행 개수, 한 줄에 10개씩 이미지 시각화, ceil 올림
    cols = N if rows < 2 else 10 # 열 개수, 행이 2개 이하면 열 개수는 이미지 개수, 행이 2개 이상이면 10개
    fig, ax = plt.subplots(rows, cols, figsize=(cols * ratio, rows * ratio), squeeze=False) # squeeze=False → 차원 축소 방지(2차원 유지)
    
    for i in range(rows):
        for j in range(cols):
            if i * 10 + j < N:
                ax[i, j].imshow(arr[i * 10 + j], cmap='gray_r')
            ax[i, j].axis('off')
    
    plt.show()
```

```python
draw_fruits(fruits[:100])
draw_fruits(fruits[100:200])
draw_fruits(fruits[200:])
```

```python
# 1차원 데이터로 펼치기
fruits_1d = fruits.reshape(300, 100 * 100)
fruits_1d.shape # (300, 10000) → 이미지 데이터를 1차원 데이터로 펼침(이해가 잘 안됨)
```

```python
# 50차원으로 차원 축소
pca = PCA(n_components=50)
pca.fit(fruits_1d)
```

```python
pca.components_.shape # (50, 10000) = (PCA로 구현한 50개의 주성분, 원본 데이터 특성 개수)
```

```python
draw_fruits(pca.components_.reshape(-1, 100, 100))
```

```python
fruits_pca = pca.transform(fruits_1d)
fruits_pca.shape # (300, 50) → 50차원으로 차원 축소
```

```python
# PCA 데이터 저장
np.save('./data/fruits_pca.npy', fruits_pca)
```

```python
# 라벨 데이터 생성
target = np.array([0] * 100 + [1] * 100 + [2] * 100)
target
```

```python
# 원본 데이터 교차검증 점수 확인
# cross_val_score: 각 CV의 점수 반환
# cross_validate: 각 CV의 학습시간, 텟트 시간, CV의 점수 반환

from sklearn.model_selection import cross_validate

lr_clf = LogisticRegression(max_iter=1000)
result = cross_validate(lr_clf, fruits_1d, target, cv=3) # 모델 객체, X 데이터, y 데이터, 교차 검증 횟수
result
# {'fit_time': array([0.30405116, 0.22480536, 0.22196174]), # 학습 시간
#  'score_time': array([0.00300431, 0.00299215, 0.00500083]), # 평가 시간
#  'test_score': array([1.  , 1.  , 0.99])} # 평가(교차 검증) 점수

# fruits_1d 는 원본 데이터를 1차원 데이터로 펼친 데이터
```

```python
# PCA 데이터 교차검증 점수 확인
result_pca = cross_validate(lr_clf, fruits_pca, target, cv=3)
result_pca
```

```python
# 원본 데이터의 교차 검증과 비교했을때, fit_time, score_time 이 줄어들고, test_score 는 동일
```

```python
fruits_inverse = pca.inverse_transform(fruits_pca) # fruits_pca 차원 축소된 데이터를 원본 데이터로 복원
fruits_inverse.shape # (300, 10000) → 원본 데이터로 복원, 하지만 이미지 배열로 우리가 3차원으로 돌려줘야 시각화 할 수 있다.
```
